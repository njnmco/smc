```python
import pacmap
import numpy as np
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()['data']
```

```python
iris[:5,:]
```

```python
m = pacmap.PaCMAP( save_tree=True)
```

```python
xhat = m.fit_transform(iris)
xhat[:5,:]
```

```python
xhat2 = m.transform(iris, None) #save_tree was true so should not need basis?
xhat2[:5,:]
```

```python
xhat3 = m.transform(iris, basis=iris.astype(np.float32)) #provide iris
xhat3[:5,:]
```

```python
xhat4 = m.transform(iris, basis=xhat) #provide predictions
xhat4[:5,:]
```

# save_tree = False

```python
m2 = pacmap.PaCMAP( save_tree=False)
```

```python
xhat = m2.fit_transform(iris)
xhat[:5,:]
```

```python
xhat2 = m2.transform(iris, None) #save_tree was false so should error?
xhat2[:5,:]
```

```python
xhat3 = m2.transform(iris, basis=iris.astype(np.float32)) 
xhat3[:5,:]
```

```python
xhat4 = m2.transform(iris, basis=xhat)
xhat4[:5,:]
```

```python
xhat5 = m2.transform(iris[:5,:], basis=iris.astype(np.float32)) 
xhat5[:5,:]
```
